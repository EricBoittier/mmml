# Run opt_mmml from notebook

Same inputs as `run_single_cutoff_test.sh`: single cutoff combination (ml=1.5, mm_on=5.0, mm_cut=1.0).

```python
import argparse
from pathlib import Path
import os

from mmml.cli.opt_mmml import run

# Run this notebook from examples/betadiket (or set base_dir to your example dir).
# Adjust dataset/checkpoint to match your environment; run_single_cutoff_test.sh uses $CHECKPOINT.
base_dir = Path.cwd()
CHECKPOINT = os.environ.get("CHECKPOINT", "path/to/checkpoint")

args = argparse.Namespace(
    dataset=Path("/home/ericb/mmml/filtered_acetone_3-8A.npz"),
    pdbfile=base_dir / "pdb" / "init-packmol.pdb",
    checkpoint=Path(CHECKPOINT),
    n_monomers=2,
    n_atoms_monomer=10,
    ml_cutoff_grid="1.5",
    mm_switch_on_grid="5.0",
    mm_cutoff_grid="1.0",
    energy_weight=1.0,
    force_weight=1.0,
    max_frames=100,
    out=base_dir / "test_cutoff_opt.json",
    out_npz=base_dir / "test_cutoff_opt.npz",
    validate=False,
    energy_catch=0.5,
    cell=None,
    ml_cutoff=1.5,
    mm_switch_on=5.0,
    mm_cutoff=1.0,
    include_mm=True,
    skip_ml_dimers=False,
    debug=False,
)

run(args)
```
