```python
R = R_initial
shift = np.zeros_like(R)
shift[:10,: ] = np.array([1,0,0]
                        )
shift

results = []
distances = []
atoms_ = []
N_scan = 200
for i in range(2, N_scan+2):
    R_ = R + shift * i * 0.081 + shift
    atoms.set_positions(R_) 
    atoms.get_forces()
    # view(atoms_minimized, viewer="x3d")
    res = atoms.calc.results
    results.append(res)
    atoms_.append(atoms.copy())
    com = np.linalg.norm(R_[:10].mean(axis=0) - R_[10:].mean(axis=0))
    oo_dist = np.linalg.norm(R_[0] - R_[10])

    distances.append(oo_dist)
```

```python
plt.plot( distances, [results[i]["model_output"].energy for i in range(N_scan)], label="E")
plt.ylim(-95,-70)
# plt.xlim(0,12)
plt.legend()
plt.xlim(0,12)
```
